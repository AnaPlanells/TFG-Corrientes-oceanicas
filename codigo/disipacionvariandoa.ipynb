```python
import numpy as np
import matplotlib.pyplot as plt

delta_P_0 = (np.load("C:\\Users\\Ana\\Desktop\\Python TFG\\eqsolvariandoa001paso0.npy") -
              np.load("C:\\Users\\Ana\\Desktop\\Python TFG\\Pi_sin1.npy")) / np.load("C:\\Users\\Ana\\Desktop\\Python TFG\\Pi_sin1.npy")

delta_P_200 = (np.load("C:\\Users\\Ana\\Desktop\\Python TFG\\eqsolvariandoa001paso200.npy") -
               np.load("C:\\Users\\Ana\\Desktop\\Python TFG\\Pi_sin1.npy")) / np.load("C:\\Users\\Ana\\Desktop\\Python TFG\\Pi_sin1.npy")


diff_delta_P = delta_P_200 - delta_P_0


plt.figure(figsize=(8, 6))
plt.contourf(diff_delta_P, cmap='bwr', levels=50)
plt.colorbar(label='ΔP (paso 200 - paso 0)')
plt.title('Diferencia en ΔP relativa entre paso 200 y paso 0')
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.tight_layout()
plt.show()
```
