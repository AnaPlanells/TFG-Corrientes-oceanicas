```python
import numpy as np
import matplotlib.pyplot as plt

# Cargar los ΔP de los pasos 10 y 200
delta_P_10 = (np.load("C:\\Users\\Ana\\Desktop\\Python TFG\\paso_101.npy") -
              np.load("C:\\Users\\Ana\\Desktop\\Python TFG\\Pi_sin1.npy")) / np.load("C:\\Users\\Ana\\Desktop\\Python TFG\\Pi_sin1.npy")

delta_P_200 = (np.load("C:\\Users\\Ana\\Desktop\\Python TFG\\paso_2001.npy") -
               np.load("C:\\Users\\Ana\\Desktop\\Python TFG\\Pi_sin1.npy")) / np.load("C:\\Users\\Ana\\Desktop\\Python TFG\\Pi_sin1.npy")

# Calcular diferencia entre ΔP de paso 200 y paso 10
diff_delta_P = delta_P_200 - delta_P_10

# Visualizar la diferencia
plt.figure(figsize=(8, 6))
plt.contourf(diff_delta_P, cmap='bwr', levels=50)
plt.colorbar(label='ΔP (paso 200 - paso 10)')
plt.title('Diferencia en ΔP relativa entre paso 200 y paso 10')
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.tight_layout()
plt.show()
```
